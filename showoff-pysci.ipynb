## Algorithms !

Let's apply some algorithms to our data !

First let's start with some dimensionality reduction and then apply clustering.

Most algorithms, distance measures, etc, are available through `scipy`.

More Machine Learning oriented bundles are available on `sklearn` (stands for scikit-learn).

### Principal Components Analysis

````python
df = df.drop(["species"], 1) # Let's drop the species
df.head()
````

````python
from sklearn.decomposition import PCA
df_norm = (df - df.mean()) / df.std() # Normalize data
pca = PCA() # Init PCA
pca_res = pca.fit_transform(df_norm.values) # apply PCA
df_pca = pd.DataFrame(pca_res, columns=pca.explained_variance_ratio_) # Make a DataFrame from the PCA results
df_pca.head() # columns name are the % of the explained variance
````

````python
# Let's keep only the first two columns which explains 95% of the variance
df_pca = df_pca.drop(df_pca.columns[3],1)
df_pca = df_pca.drop(df_pca.columns[2],1)
````

````python
# Let's round columns name
pca_res = pd.DataFrame(df_pca.values, columns=[round(x, 2) for x in df_pca.columns])
pca_res.head()
````

Cool, so now we can vizualize our data in two dimensions !

````python
plt.scatter(pca_res[0.73], pca_res[0.23])
plt.show()
````

Now we can apply some clustering

### K-means (clustering)

````python
from scipy import cluster
def myKmeans(df, nb_clusters):
    centroids, _ = cluster.vq.kmeans(df.values, nb_clusters, iter=100)
    idx, _ = cluster.vq.vq(df.values, centroids)
    return idx
````

````python
km = myKmeans(pca_res, 3)
````

````python
colors = "bgrcmyk" # each letter is a different colour
for c in set(km):
    plt.scatter(pca_res[0.73][iris.target == c], pca_res[0.23][iris.target == c], c=colors[c])
plt.show()
````

## Networks !

With [`networkx`](https://networkx.github.io/).
Lot of algorithms, and some visualizations.

````python
import networkx as nx
import matplotlib.pyplot as plt

G=nx.Graph()

G.add_edge('a','b',weight=0.6)
G.add_edge('a','c',weight=0.2)
G.add_edge('c','d',weight=0.1)
G.add_edge('c','e',weight=0.7)
G.add_edge('c','f',weight=0.9)
G.add_edge('a','d',weight=0.3)

elarge=[(u,v) for (u,v,d) in G.edges(data=True) if d['weight'] >0.5]
esmall=[(u,v) for (u,v,d) in G.edges(data=True) if d['weight'] <=0.5]

pos=nx.spring_layout(G) # positions for all nodes

# nodes
nx.draw_networkx_nodes(G,pos,node_size=700)

# edges
nx.draw_networkx_edges(G,pos,edgelist=elarge,
                    width=6)
nx.draw_networkx_edges(G,pos,edgelist=esmall,
                    width=6,alpha=0.5,edge_color='b',style='dashed')

# labels
nx.draw_networkx_labels(G,pos,font_size=20,font_family='sans-serif')

plt.axis('off')
plt.show() # display
````
